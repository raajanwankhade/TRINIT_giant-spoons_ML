```python
from datasets import Dataset, load_dataset
import pandas as pd
from nltk.translate.bleu_score import sentence_bleu
```

```python
test_dataset = load_dataset("arampacha/rsicd",split="test")
test_dataset
```

```python
df = pd.read_csv("/kaggle/input/evals-rsicd/test_captions.csv")
df = df.drop(["Unnamed: 0"], axis = 1)
df
```

```python
import random
```

```python
random_captions = {}
for i, row in enumerate(test_dataset):
    image_filename = row["filename"]
    captions = list(set(row["captions"]))
    random_caption = random.choice(captions)
    random_captions[image_filename] = random_caption
```

```python
len(random_captions)
```

```python
bleu_scores = []
for index, row in df.iterrows():
    image_filename = row["filenames"]
    generated_caption = row["generated_captions"]
    
    if image_filename in random_captions:
        reference = random_captions[image_filename]
        bleu_score = sentence_bleu([reference.split()], generated_caption.split())
        bleu_scores.append(bleu_score)
```

```python
print("Average BLEU Score for Test Set:", sum(bleu_scores) / len(bleu_scores))
```
